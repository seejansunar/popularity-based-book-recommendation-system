## Popularity based Recommendar System

##### Top 50 books which has got more than average ratings but atleast they should have 250+ votes.

```python
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.head()
```

```python
num_rating_df = ratings_with_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating_df.sort_values('num_ratings', ascending=False)
```

```python
avg_rating_df = ratings_with_books.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'}, inplace=True)
avg_rating_df.sort_values('avg_ratings', ascending=False)
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title').sort_values('num_ratings', ascending=False)
popular_df
```

```python
popular_df = popular_df[popular_df["num_ratings"]>=250].sort_values("avg_ratings", ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')
```

```python
popular_df
```

## Collaborative filtering based Recommendar System

```python
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
x = ratings_with_books.groupby('User-ID').count()['Book-Rating']>200
top_users = x[x].index
```

```python
top_users
```

```python
filtered_ratings = ratings_with_books[ratings_with_books["User-ID"].isin(top_users)]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings["Book-Title"].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    book_index = np.where(pt.index==book_name)[0][0]
    similar_books  = sorted(list(enumerate(similarity_scores[book_index])), key=lambda x:x[1], reverse=True)[1:6]
    for item in similar_books:
        print(pt.index[item[0]])
```

```python
recommend('Message in a Bottle')
```
